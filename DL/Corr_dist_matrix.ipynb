```python
import distcorr as dis
import numpy as np
import seaborn as sns
```

```python
import seaborn as sns
def dist_corr_matrix(points, val=True):
    """
    a = [1,2,3,4,5]
    b = np.array([1,2,9,4,4])
    c=np.array([3,6,-2,6,9])
    points=[a,b,c]
    """
    points=np.atleast_1d(points)
    corre=np.arange(float(len(points))**2).reshape(len(points),len(points))
    for i in range(len(points)):
        for j in range(len(points)):
            corre[i,j]=float(dis.dist_corr(points[i], points[j]))
    sns.heatmap(corre, annot=val)
```

```python
points=[]

for i in range(15): # 15 is the number of 3D points
    points.append(10*np.random.randn(3)) # 3D points
print(points)

dist_corr_matrix(points)
```
